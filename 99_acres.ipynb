```python
from selenium import webdriver

driver = webdriver.Firefox()
driver.get("https://www.magicbricks.com/localities-in-bangalore")

# Get the height of the initial page
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to the bottom of the page
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
# Use BeautifulSoup to extract data from the page
soup = BeautifulSoup(driver.page_source, 'html.parser')
property_names = []


lists = soup.find_all( 'div' ,class_ = 'loc-card' )
for list1 in lists:
    name = list1.find( 'a' ,class_ =  'loc-card__title').text
    property_names.append(name)
print(len(property_names))
```
